## **Module H1: Ordinary Differential Equations**

**22.1: ODE Initial Value Problem Statement**

A differential equation is a relationship between a function, ùëì(ùë•), its independent variable, ùë•, and any number of its derivatives. An **ordinary differential equation** or **ODE** is a differential equation where the independent variable, and therefore also the derivatives, is in one dimension. 

$$
F(x, f(x), \frac{df(x)}{dx}, \frac{d^2f(x)}{dx^2},\frac{d^3f(x)}{dx^3}, ..., \frac{d^{n-1}f(x)}{dx^{n-1}}) = \frac{d^nf(x)}{dx^n},
$$

where ùêπ is an arbitrary function that incorporates one or all of the input arguments, and $n$ is the order of the differential equation. This equation is referred to as an **$n^{th}$ order ODE**.

A general solution to a differential equation is a $g(x)$ that satisfies the differential equation. Although there are usually many solutions to a differential equation, they are still hard to find. For an ODE of order $n$, a particular solution is a $p(x)$ that satisfies the differential equation and n explicitly known values of the solution, or its derivatives, at certain points. Generally stated, $p(x)$ must satisfy the differential equation and $p^{(j)}(x_\mathrm{i})=p_\mathrm{i}$, where $p^{(j)}$ is the $j^{th}$ derivative of $p$, for $n$ triplets, $(j,x_\mathrm{i}, p_\mathrm{i})$. For the purpose of this text, we refer to the particular solution simply as the **solution**.
A common set of known values for an ODE solution is the **initial value**.For an ODE of order $n$, the initial value is a known value for the $0^{th}$ to $(ùëõ‚àí1)^{th}$ derivatives at $x=0,f(0), f^{(1)}(0), f^{(2)}(0),..., f^{(n-1)}(0)$. Finding a solution to an ODE given an initial value is called the **initial value problem**. 


**22.2: Reduction of Order**

To reduce the order of a differential equation, consider a vector, $S(t), which is the **state** of the system as a function of time. In general, the state of a system is a collection of all the dependent variables that are relevant to the behavior of the system. For the following ODE:

$$
F(x, f(x), \frac{df(x)}{dx}, \frac{d^2f(x)}{dx^2},\frac{d^3f(x)}{dx^3}, ..., \frac{d^{n-1}f(x)}{dx^{n-1}}) = \frac{d^nf(x)}{dx^n},
$$

for initial value problems, it is useful to take the state to be:


\begin{align}
S(t) = 
\left[\begin{array}{c}
f(t) \\ f^{(1)}(t) \\ f^{(2)}(t) \\ f^{(3)}(t) \\ ... \\ f^{(n-1)}(t)
\end{array}\right]
& 
\end{align}

Then derivative of the state is 

\begin{align}
\frac{dS(t)}{dt} = 
\left[\begin{array}{c}
f^{(1)}(t) \\ f^{(2)}(t) \\ f^{(3)}(t) \\f^{(4)}(t) \\ ... \\ f^{(n)}(t)
\end{array}\right]
& = \left[\begin{array}{c}
f^{(1)}(t) \\ f^{(2)}(t) \\ f^{(3)}(t) \\f^{(4)}(t) \\ ... \\ F(t, f(t), f^{(1)}(t), ..., f^{(n-1)}(t)
\end{array}\right]
& = \left[\begin{array}{c}
S_\mathrm{2}(t) \\ S_\mathrm{3}(t) \\ S_\mathrm{4}(t) \\ S_\mathrm{5}(t) \\ ... \\ F(t, S_\mathrm{1}(t), S_\mathrm{2}(t), ..., S_\mathrm{n-1}(t)
\end{array}\right]
&
\end{align}

where $S_\mathrm{i}(t)$ is the $i^{th}$ element of $S(t)$.
With the state written in this way, $\frac{dS(t)}{dt}$ can be written using only $S(t)$ (i.e., no$f(t)$) or its derivatives. In particular, $\frac{dS(t)}{dt} = ‚Ñ±(t, S(t))$, where $‚Ñ±$ is a function that appropriately assembles the vector describing the derivative of the state. This equation is in the form of a first-order differential equation in $S$.

**22.3: The Euler Method**

Let $\frac{ùëëùëÜ(ùë°)}{ùëëùë°}=ùêπ(ùë°,ùëÜ(ùë°))$ be an explicitly defined first order ODE. That is, $ùêπ $ is a function that returns the derivative, or change, of a state given a time and state value. Also, let ùë° be a numerical grid of the interval $[t_\mathrm{0},t_\mathrm{f}]$ with spacing $‚Ñé$. Without loss of generality, we assume that $t_\mathrm{0}$=0, and that $t_\mathrm{f}=ùëÅ‚Ñé$ for some positive integer, $ùëÅ$.

The linear approximation of $ùëÜ(ùë°)$ around $t_\mathrm{j} at $t_\mathrm{j+1} is

$$
ùëÜ(t_\mathrm{j+1})=ùëÜ(t_\mathrm{j})+(t_\mathrm{j+1}‚àít_\mathrm{j})\frac{dS(t_\mathrm{j})}{dt}
$$

which can also be written

$$
ùëÜ(t_\mathrm{j+1})=ùëÜ(t_\mathrm{j})+hF(t_\mathrm{j}, S(t_\mathrm{j}))
$$
This formula is called the **Explicit Euler Formula**, and it allows us to compute an approximation for the state at $S(t_\mathrm{j+1})$ given the state at $S(t_\mathrm{j})$. Starting from a given initial value of $S_\mathrm{0} = S(t_\mathrm{0})$, we can use this formula to integrate the states up to $S(t_\mathrm{f})$; these $S(t)$ values are then an approximation for the solution of the differential equation. 

Assume we are given a function $ùêπ(ùë°,ùëÜ(ùë°))$ that computes $\frac{ùëëùëÜ(ùë°)}{ùëëùë°}$, a numerical grid, $ùë°$, of the interval, $[t_\mathrm{0},t_\mathrm{f}]$ and an initial state value $S_\mathrm{0} = S(t_\mathrm{0})$. We can compute $S(t_\mathrm{j})$ for every $t_\mathrm{j}$ in $t$ using the following steps.

1. Store $S_\mathrm{0} = S(t_\mathrm{0})$ in an array, $S$.

2. Compute $S(t_\mathrm{1}) = S_\mathrm{0} + hF(t_\mathrm{0}, S_\mathrm{0})$.

3. Store $S_\mathrm{1} = S(t_\mathrm{1})$ in  $S$.

4. Compute Compute $S(t_\mathrm{2}) = S_\mathrm{1} + hF(t_\mathrm{1}, S_\mathrm{1})$.

5. Store $S_\mathrm{2} = S(t_\mathrm{2})$ in  $S$.

6. ...

7. Compute $S(t_\mathrm{f}) = S_\mathrm{f-1} + hF(t_\mathrm{f-1}, S_\mathrm{f-1})$..

8. Store $S_\mathrm{f} = S(t_\mathrm{f})$ in  $S$.

9. $ùëÜ$ is an approximation of the solution to the initial value problem.

When using a method with this structure, we say the method **integrates** the solution of the ODE.



**Example:** The differential equation $\frac{df(t)}{dt} = e^{-t}$  with initial condition $f_\mathrm{0} =-1$  has the exact solution $f(t) = - e^{-t}$. Approximate the solution to this initial value problem between 0 and 1 in increments of 0.1 using the Explicity Euler Formula. Plot the difference between the approximated solution and the exact solution.



```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

In the above figure, we can see each dot is one approximation based on the previous dot in a linear fashion. From the initial value, we can eventually get an approximation of the solution on the numerical grid. If we repeat the process for $‚Ñé=0.01$, we get a better approximation for the solution:

```python
h = 0.01 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'b--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

The **Implicit Euler Formula** can be derived by taking the linear approximation of $ùëÜ(ùë°)$ around $t_\mathrm{j+1}$ and computing it at $t_\mathrm{j}$:
$$
ùëÜ(t_\mathrm{j+1})=ùëÜ(t_\mathrm{j})+hF(t_\mathrm{j+1}, S(t_\mathrm{j+1}))
$$

The**Trapezoidal Formula** is the average of the Explicit and Implicit Euler Formulas:

$$
ùëÜ(t_\mathrm{j+1})=ùëÜ(t_\mathrm{j})+\frac{h}{2}(F(t_\mathrm{j}, S(t_\mathrm{j})) + F(t_\mathrm{j+1}, S(t_\mathrm{j+1})))
$$

###**Exercises:**

Define a function odeEuler(f, y0, t) to approximate the solution of y'=f(t,y) using Explicit Euler Formular.\
f: is the right-hand side of the differential equation y'= f(t,y); \
y0: initial value at t0, y(t0) = y0; \
t: 1D NumPy array of t values where we approximate y values. Time step at each iteration is given by $t_\mathrm{n+1}-t_\mathrm{n}$.\
The function returns y : 1D NumPy array. Approximation y[n] of the solution y(t_n) computed by Euler's method.

```python
import numpy as np
import matplotlib.pyplot as plt

def odeEuler(f,y0,t):
    
    y = np.zeros(len(t))
    y[0] = y0
    for n in range(0,len(t)-1):
        y[n+1] = y[n] + f(y[n],t[n])*(t[n+1] - t[n])
    return y
```

**(1) Exponential equation:**Approximate solutions of $y'=y$ for $t\in [0,2]$. Compare with the exact solution $y=e^t$.

```python
t = np.linspace(0,2,21)
y0 = 1
f = lambda y,t: y
y = odeEuler(f,y0,t)
y_true = np.exp(t)
plt.figure(figsize = (12, 8))
plt.plot(t,y,'b.-',t,y_true,'r-')
plt.legend(['Euler','True'])
plt.axis([0,2,0,9])
plt.grid(True)
plt.title("Solution of $y'=y , y(0)=1$")
plt.show()
```

**(2) Non-Linear Equation:**
Plot an approximation of $y'=y^2$ with $y_\mathrm{0}=-1$. Compare with the exact solution $y=\frac{-1}{t+1}$.


```python
t = np.linspace(0,5,16)
y0 = -1
f = lambda y,t: y**2
y = odeEuler(f,y0,t)
t_true = np.linspace(0,5,100)
y_true = -1/(t_true + 1)
plt.figure(figsize = (12, 8))
plt.plot(t,y,'r.-',t_true,y_true)
plt.legend(['Euler','True'])
plt.grid(True)
plt.axis([0,5,-1,0])
plt.title("Solution of $y'=y^2 , y(0)=1$")
plt.show()
```

**(3) logistic equation:** plot an approximation for logistic equztion $\frac{dP}{dt}=1.1*P(1-\frac{P}{20})$ for $t\in [0,20]$ with $P_\mathrm{0}=10$. Compare with the exact solution $P(t) = \frac{20*P_\mathrm{0}*e^{1.1t}}{20+P_\mathrm{0}(e^{1.1t}-1)}$.

```python
t = np.linspace(0, 20, 41)
P0 = 10
f = lambda P,t: 1.1*P*(1-P/20)

P = odeEuler(f, P0, t)

t_true = np.linspace(0, 20, 2001)
p_true= 20*P0*np.exp(1.1*t_true)/(20+P0*(np.exp(1.1*t_true)-1))

plt.figure(figsize = (12, 8))
plt.plot(t,P,'r.-')
plt.plot(t_true,p_true, 'b')
plt.legend(['Euler','True'])
plt.grid(True)
plt.axis([-1,21,8,22])
plt.title("Solution of $P'=1.1P(1-P/20), P(0)=10$")
plt.show()




```
