#### **3.2.1 Limits and continuity.** SymPy can compute symbolic limits with the **limit** function. The syntax to compute {lim f(x), x -> x0} is **limit(f(x),  x,  x0)**.


```python
# import library

import sympy as sym

# pythonic math expressios: add spaces, use single quotes and lowercases
# declaring variables as symbols
x, n = sym.symbols('x, n')

# define a function f1, and print the limit while x is approaching 0
f1 = sym.sin(x) / x
print('while x is approaching 0, the limit of sin(x)/x is: ', sym.limit(f1, x, 0))

# define another function f2, print the limit of the function whils n is approaching infinity
f2 = n**2/sym.exp(n)
print('while n is approaching infinity, the limiti of n^2/exp(n) is: ', 
      sym.limit(f2, n, sym.oo)) # sym.oo represent the mathematical infinity
```

#### **Derivative Rules for basic functions**

##### (1) derivative of powers: f(x) = x^n, f'(x) = n*x^(n-1)

##### (2) Exponential and logrithmic functions: 
##### f(x) = e^x, f'(x) = e^x; 
##### f(x) = e^(c*x), f'(x) = c*e^(c*x), c is a constant; 
##### f(x) = a^x, f'(x) = a^x * ln(a), a>0;
##### f(x) = ln(x), f'(x) = 1/x, x>0;


##### (3) sin and cos functions:
##### f(x) = sin(x), f'(x) = cos(x)
##### f(x) = cos(x), f'(x) = -sin(x)

#### **Rules for combined functions**

##### (1) constant rules: if f(x) is a constant, then f'(x) = 0

##### (2) sum rule: [a*f(x) + b*g(x)]' = a*f'(x) + b*g'(x)

##### (3) Product rule: [f(x)g(x)]' = f'(x)g(x) + f(x)g'(x)

##### (4) Quotient rule: [f(x)/g(x)]' = [f'(x)g(x)-f(x)g'(x)]/g^2

##### (5) Chain rule: if h(x) = f(g(x)), then h'(x) = f'(g(x)).g'(x)

#### **3.2.2 Derivatives**: To take the derivative, use the **diff** function.

```python
# declaring variables
x, y , z= sym.symbols('x, y, z')

# define a function

f = sym.cos(x) + sym.exp(5*x) + sym.log(x)

# diff(function, independent variable, number of derivatives)

f_prime1 = sym.diff(f, x, 1)
print("1st derivative f'(x) = ", f_prime1)

f_prime2 = sym.diff(f, x, 2)
print("2nd derivative f''(x) = ", f_prime2)

f_prime3 = sym.diff(f, x, 3)
print("3rd derivative f'''(x) = ", f_prime3)

```

#### **3.2.3 Taylor's Theorem**: Taylor's theorem gives an approximation of a differentiable function around a given point by a polynomial.

##### To compute the expansion of 洧녭(洧논) around the point 洧논=洧논0 terms of order 洧논^洧녵, use f(x).series(x, x0, n). x0 and n can be omitted, in which case the defaults x0=0 and n=6 will be used.


```python
# calculate the taylor expension of cos(x) for different terms.
x = sym.symbols('x')
f = sym.cos(x)
f2 = f.series(x, 0, 2) 
f3 = f.series(x, 0, 3) 
f4 = f.series(x, 0, 4)
f5 = f.series(x, 0, 5).removeO()
f6 = f.series(x, 0, 6).removeO() #removeO()removes the 0(x^n) term
f7 = f.series(x, 0, 7).removeO()
f9 = f.series(x, 0, 9).removeO()
f11 = f.series(x, 0, 11).removeO()
print('f2 = ', f2, '\nf3 = ', f3, '\nf4 = ', f4, '\nf5 = ', f5, '\nf6 = ', f6, '\nf7 = ', f7, '\nf9 = ', f9, '\nf11 = ', f11)
```

```python
# plot cos(x) and its Taypor expansion at terms 4, 7, 9 and 11

import matplotlib.pyplot as plt
import numpy as np

#plt.style.use('seaborn-poster')
plt.figure(figsize = (10, 4))
x = np.arange(-5, 5, 0.01)
f = np.cos(x) 
f4 =  1 - x**2/2 
f7 =  -x**6/720 + x**4/24 - x**2/2 + 1 
f9 =  x**8/40320 - x**6/720 + x**4/24 - x**2/2 + 1 
f11 = -x**10/3628800 + x**8/40320 - x**6/720 + x**4/24 - x**2/2 + 1
plt.plot(x, f, '-r', label = 'cos(x)')
plt.plot(x,f4, '-b', label = 'f4')
plt.plot(x,f7, 'black', label = 'f7')
plt.plot(x,f9, 'purple', label = 'f9')
plt.plot(x,f11, 'orange', label = 'f11')
plt.title('cos(x) and Taylor Expansion of cos(x) at n = 4, 7, 9 and 11')
plt.xlabel('x')
plt.ylabel('f')
plt.ylim([-2, 2])
plt.legend()
plt.grid()
plt.show()
```

#### **3.3 Unconstrained Optimization**

##### Definition 3.3.1 (Global minimizer):
Let $f : \mathbb{R}^d\to\mathbb{R}$. The point $\textbf{x}^*\in\mathbb{ R}^d$
is a global minimizer of $f$ over $\mathbb{R}^d$ if
$$f(\textbf{x}) \geq f(\textbf{x}^*), \forall \textbf{x} \in \mathbb{ R}^d.$$

##### Definition 3.3.2 (Local minimizer):
Let $f : \mathbb{R}^d\to\mathbb{R}$. The point $\textbf{x}^*\in\mathbb{ R}^d$
is a global minimizer of $f$ over $\mathbb{R}^d$ if there is $\delta >0$ such that 
$$f(\textbf{x})\geq f(\textbf{x}^*), \forall \textbf{x}\in B_\delta(\textbf{x}^*)\backslash \{\textbf{x}^*\}.$$
If the inequality is strict, we say that $\textbf{x}^*$ is a strict local minimizer.

##### **(1) SciPy optimize** provides functions for minimizing (or maximizing) objective functions, possibly subject to constraints. It includes solvers for nonlinear problems (with support for both local and global optimization algorithms), linear programing, constrained and nonlinear least-squares, root finding, and curve fitting.

##### **Scalar functions optimization**: minimize_scalar(fun[, bracket, bounds, 뵣) -- Minimization of scalar function of **one variable.**


```python
from scipy import optimize
import numpy as np

f = lambda x: x**2 -12*x + 20  # define a function

result = optimize.minimize_scalar(f)
if result.success:    # if solver was successful, do the following
  x_min = result.x    # the x value at the minimum
  y_min = result.fun  # the y value at the minimum
  print("Success!")
  print(f"x = {x_min} \ny = {y_min} ")

else:
  print('Sorry, could not find the minimum!')
```
