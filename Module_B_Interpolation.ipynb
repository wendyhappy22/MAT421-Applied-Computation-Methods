### Linear Interpolation
###### xi<x<x(i+1),  y_hat = yi +[y(i+1)-yi]*(x-xi)/[x(i+1)-xi].
###### use scipy's function interp1d

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = [0, 1, 2]
y = [1, 3, 2]
f = interp1d(x, y)
y_hat = f(1.5)
print(y_hat)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x, y, '-ob')
plt.plot(1.5, y_hat, 'ro')
plt.title('Linear Interolation at x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Cubic Spline Interpolation
###### assume that the  points (xi, yi)and [x(i+1), y(i+1)] are joined by a cubic polynomial Si(x) = ai*x^3 + bi*x^2 + ci*x + di, xi <= x <=x(i+1), i=1,.....,(n-1).
###### In Python, scipy's function CubicSpline can be used to perform cubic spline interpolation.

```python
# use CubicSpline to plot the cubic cpline interpolation of the data set x = [0,1,2] and y = [1,3,2] for 0<=x<=2
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
x = [0, 1, 2]
y = [1, 3, 2]
# use bc_type = 'natural' to add the constraints (assume the second derivatives are zero at the endpoints)
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)
plt.figure(figsize = (10, 8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Lagrange Polynomial Interpolation
###### Largrange polynomial L(x) finds a single polynomial that goes through all the data points. L(xi) = yi.

```python
# using lagrange from scipy
from scipy.interpolate import lagrange

x = [0, 1, 2]
y = [1, 3, 2]
f = lagrange(x, y)

x_new = np.arange(-1.0, 3.1, 0.1)

fig = plt.figure(figsize = (10, 8))
plt.plot(x_new, f(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomials')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

###### (1) Write a function my_interp_plotter(x, y, X, option), where x and y are arrays containing experimental data points, and X is an array containing the coordinates for which an interpolation is desired. The input argument option should be a string, either ‘linear’, ‘spline’, or ‘nearest’. Your function should produce a plot of the data points (𝑥,𝑦) marked as red circles. and the points (𝑋,𝑌), where X is the input and Y is the interpolation at the points contained in X defined by the input argument specified by option. The points (𝑋,𝑌) should be connected by a blue line. Be sure to include title, axis labels, and a legend. Hint: You should use interp1d from scipy, and checkout the kind option.

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

def my_interp_plotter(x, y, X, option):
  
  plt.plot(x, y, 'ro', label = 'data point')
  
  if option == 'nearest':
    f_nearest = interp1d(x, y, kind = 'nearest') #takes the given data points, fits the model specified and return the fitted function
    plt.plot(X, f_nearest(X), 'b', label = 'interpolation')
    plt.title('nearest interpolation of data')

  elif option == 'linear':
    f_linear = interp1d(x, y, kind = 'linear')
    plt.plot(X, f_linear(X), 'b', label = 'interpolation')
    plt.title('linear interpolation of data')

  elif option == 'cubic':
    f_cubic = interp1d(x, y, kind = 'cubic')
    plt.plot(X, f_cubic(X), 'b', label = 'interpolation')
    plt.title('cubic interpolation of data')
  
  else:
    print ("please select from \"nearest\", \"linear\", and \"cubic\" as options to interpolate ")

  plt.xlabel('x')
  plt.ylabel('y')
  plt.legend()
  plt.show()
```

###### (2) Write a function my_D_cubic_spline(x, y, X, D), where the output Y is the cubic spline interpolation at X taken from the data points contained in x and y. However, instead of the standard pinned endpoint conditions (i.e., 𝑆″1(𝑥1)=0 and 𝑆″𝑛−1(𝑥𝑛)=0) you should use the endpoint conditions 𝑆′1(𝑥1)=𝐷 and 𝑆′𝑛−1(𝑥𝑛)=𝐷 (i.e., the slopes of the interpolating polynomials at the endpoints is 𝐷).



```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
plt.figure(figsize = (10,8))

def my_D_cubic_spline(x, y, X, D):
  # use bc_type = ((1,D), (1,D)) adds the constraints that the first derivate at curve ends are of a value D
  f = CubicSpline(x, y, bc_type = ((1, D),(1, D)))
  Y = f(X)
  return Y
```

###### (3) Write a function my_lagrange(x, y, X), where the output Y is the Lagrange interpolation of the data points contained in x and y computed at X. Hint: Use a nested for-loop, where the inner for-loop computes the product for the Lagrange basis polynomial and the outer loop computes the sum for the Lagrange polynomial. Don’t use the existing lagrange function from scipy

```python
import numpy as np
import matplotlib.pyplot as plt

def my_lagrange(x, y, X):
  sum = 0  # initial value of sum is 0
  for i in range(len(x)):
    p = 1  # p means product, the initial value of the product should be 1, inner loop to calculate P(X)
    for j in range(len(x)):
      if j != i:
        p *= (X-x[j])/(x[i] - x[j])
    sum += y[i]*p   # outer loop to calculate the sum of polynomials
  return sum
  
```

```python
#test case
x = [0, 1, 2, 3, 4]
y = [2, 1, 3, 5, 1]

X = np.linspace(0, 4, 101)

plt.figure(figsize = (10,8 ))
plt.plot(X, my_lagrange(x, y, X), 'b', label = 'interpolation')
plt.plot(x, y, 'ro', label = 'data points')

plt.xlabel('x')
plt.ylabel('y')

plt.title('Lagrange Interpolation of Data Points')
plt.legend()
plt.show()
```

###### (4) Fit the data x = [0, 1, 2, 3, 4], y = [2, 1, 3, 5, 1] using Newton’s polynomial interpolation

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

%matplotlib inline
```

```python
def divided_diff(x, y):
    '''
    function to calculate the divided
    differences table
    '''
    n = len(y)
    coef = np.zeros([n, n])
    # the first column is y
    coef[:,0] = y
    
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
           (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
            
    return coef

def newton_poly(coef, x_data, x):
    '''
    evaluate the newton polynomial 
    at x
    '''
    n = len(x_data) - 1 
    p = coef[n]
    for k in range(1,n+1):
        p = coef[n-k] + (x -x_data[n-k])*p
    return p
```

```python
x = np.array([0, 1, 2, 3, 4])
y = np.array([2, 1, 3, 5, 1])
# get the divided difference coef
a_s = divided_diff(x, y)[0, :]

# evaluate on new data points and draw a smooth line when x in (0, 4)
x_new = np.arange(0, 4.1, .1)
y_new = newton_poly(a_s, x, x_new)

plt.figure(figsize = (12, 8))
plt.plot(x, y, 'bo')
plt.plot(x_new, y_new)
```
